Os dados foram obtidos diretamente de uma p√°gina da Wikip√©dia, que cont√©m uma lista de pa√≠ses por popula√ß√£o. Para isso, utilizamos o m√©todo pd.read_html() da biblioteca Pandas, que permite extrair tabelas HTML e convert√™-las em DataFrames.
